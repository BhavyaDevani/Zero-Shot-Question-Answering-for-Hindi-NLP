```python
# Step 1: Load Pretrained Multilingual Model (e.g., mBERT or XLM-R)
model_name = "bert-base-uncased"  # BERT model for English text
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
#Step 2: Load SQuAD Dataset
train_squad = load_dataset("squad", split="train")
eval_squad = load_dataset("squad", split="validation")
```

```python
# Preprocess Function for Tokenization
def preprocess_function(examples):
    inputs = tokenizer(
        examples["question"],
        examples["context"],
        max_length=384,
        truncation="only_second",
        return_offsets_mapping=True,
        padding="max_length",
    )

    offset_mapping = inputs.pop("offset_mapping")
    answers = examples["answers"]
    start_positions = []
    end_positions = []

    for i, offset in enumerate(offset_mapping):
        answer = answers[i]["text"][0]
        start_char = answers[i]["answer_start"][0]
        end_char = start_char + len(answer)

        sequence_ids = inputs.sequence_ids(i)

        # Find the start and end of the context
        context_start = sequence_ids.index(1)
        context_end = len(sequence_ids) - sequence_ids[::-1].index(1)

        # Initialize start and end tokens as 0
        start_token = 0
        end_token = 0

        # Find the start token index for answer span within context
        for j, (start, end) in enumerate(offset[context_start:context_end], start=context_start):
            if start <= start_char < end:
                start_token = j
            if start < end_char <= end:
                end_token = j
                break

        start_positions.append(start_token)
        end_positions.append(end_token)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs
```

```python
# Apply Tokenization to Training and Evaluation Data
tokenized_train_squad = train_squad.map(preprocess_function, batched=True, remove_columns=train_squad.column_names)
tokenized_eval_squad = eval_squad.map(preprocess_function, batched=True, remove_columns=eval_squad.column_names)
```

```python
from transformers import TrainingArguments, Trainer
import numpy as np

# Define F1 and Exact Match metrics
def squad_metrics(predictions, references):
    f1 = 0
    exact_match = 0
    total = len(references)

    for pred, ref in zip(predictions, references):
        # Convert prediction and reference to strings for comparison
        pred_tokens = pred.split()
        ref_tokens = ref.split()

        # Exact Match
        if pred == ref:
            exact_match += 1

        # F1 Score Calculation
        common_tokens = set(pred_tokens) & set(ref_tokens)
        num_common = len(common_tokens)
        if num_common == 0:
            continue
        precision = num_common / len(pred_tokens)
        recall = num_common / len(ref_tokens)
        f1 += 2 * (precision * recall) / (precision + recall)

    # Normalize metrics
    return {
        "f1": f1 / total,
        "exact_match": exact_match / total,
    }
```

```python
# Define the compute_metrics function for Trainer
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=-1)  # Convert logits to class indices

    # Post-process predictions and labels to convert IDs back to text
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Compute SQuAD-style metrics
    metrics = squad_metrics(decoded_preds, decoded_labels)
    return metrics
```

```python
# Define Training Arguments
training_args = TrainingArguments(
    output_dir="./results",  # Output directory for model checkpoints
    evaluation_strategy="epoch",  # Evaluate at the end of each epoch
    learning_rate=3e-5,
    per_device_train_batch_size=8,  # Adjusted batch size for Colab GPU
    per_device_eval_batch_size=8,  # Evaluation batch size
    num_train_epochs=1,  # Number of training epochs
    save_strategy="epoch",  # Save checkpoint at end of each epoch
    save_total_limit=2,  # Keep only the last 2 checkpoints
    load_best_model_at_end=True,  # Load the best model at the end based on eval loss
    logging_dir='./logs',  # Directory for logs
    logging_steps=10,  # Log every 10 steps
)

# Define Trainer with compute_metrics
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_squad,
    eval_dataset=tokenized_eval_squad,
    compute_metrics=compute_metrics,  # Pass the compute_metrics function
)

# Train the model
trainer.train()
```

```python
# Evaluate the Model after Training
eval_results = trainer.evaluate()

# Print Evaluation Results
print("Evaluation Results:")
print(f"Evaluation Loss: {eval_results['eval_loss']:.4f}")
print(f"F1 Score: {eval_results['eval_f1']:.4f}")
print(f"Exact Match (EM): {eval_results['eval_exact_match']:.4f}")
```

```python
# Step 6: Zero-shot Evaluation on Hindi using XQuAD or MLQA
xquad = load_dataset("xquad", "xquad.hi", split="validation")  # Using Hindi portion of XQuAD
```

```python
# Process XQuAD similarly
tokenized_xquad = xquad.map(preprocess_function, batched=True, remove_columns=xquad.column_names)
```

```python
# Step 7: Define Evaluation Metric
!pip install evaluate
import evaluate

metric = evaluate.load("squad")

def compute_metrics(p):
    start_pred = p.predictions[0].argmax(dim=-1)
    end_pred = p.predictions[1].argmax(dim=-1)

    # Compute the metrics and access the 'f1' score from the result
    computed_metrics = metric.compute(predictions=[{"prediction_text": tokenizer.decode(tokenized_xquad[i]["input_ids"][start_pred[i]: end_pred[i]])} for i in range(len(start_pred))], references=[{"answers": answer} for answer in xquad["answers"]])

    # Return the computed metrics including F1
    return computed_metrics
```

```python
# Evaluate
results = trainer.evaluate(eval_dataset=tokenized_xquad)

print(f"Evaluation results on Hindi: {results}")
```
